**Importing the necessary libraries and dependencies necessary for the project**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection and Data Processing**

```python
# Loading the dataset to pandas dataframe
sonar_data = pd.read_csv("/content/sonar_data.csv", header=None)
```

```python
sonar_data.head()
```

```python
# seperating data and labels
X = sonar_data.drop(columns = 60, axis=0)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

**Training and Test Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

**Model Training using Logisitic Regression**

```python
model = LogisticRegression()
```

```python
# Training the logisitic Regression model with training data
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data", training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on testing data", test_data_accuracy)
```
